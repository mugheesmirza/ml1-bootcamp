#### Further improvement using Templates

```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate

template = """
Answer the following question.

Here is the conversation history: {context}

Question: {question}

Answer:
"""

model = OllamaLLM(model = "llama3")
prompt = ChatPromptTemplate.from_template(template)
chain = prompt | model

def converstaion():
    context = ""
    print("Welcome to ChatBot! Type 'exit' to quit.")

    while True:
        question = input("You: ")
        print(f"You: {question}")
        if question.lower() == "exit":
            break

        response = chain.invoke({
            "question": question,
            "context": context
            })
        
        print(f"ChatBot: {response}")

        context += f"\nUser: {question}\nChatBot: {response}"
if __name__ == "__main__":
    converstaion()
```

```python
import os
import fitz  # PyMuPDF
import faiss
from sentence_transformers import SentenceTransformer
import google.generativeai as genai
import gradio as gr
import numpy as np
```

```python
# Configure the API key
genai.configure(api_key=os.environ['API_KEY'])

# Extract text from PDF
def extract_text_from_pdf(pdf_path):
    document = fitz.open(pdf_path)
    text = ""
    for page_num in range(document.page_count):
        page = document.load_page(page_num)
        text += page.get_text()
    return text
```

```python
def handle_pdf(pdf_path):

    pdf_text = extract_text_from_pdf(pdf_path.name)

    # Encode and index the extracted text
    sentence_model = SentenceTransformer('all-MiniLM-L6-v2')
    passages = pdf_text.split("\n\n")
    embeddings = sentence_model.encode(passages, convert_to_tensor=True)
    print(embeddings)
    
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(embeddings.cpu().numpy())

def retrieve_passages(query, k=5):
    sentence_model = SentenceTransformer('all-MiniLM-L6-v2')
    query_embedding = sentence_model.encode(query, convert_to_tensor=True)
    query_embedding = np.expand_dims(query_embedding, axis=0)  # Ensure it is 2D
    distances, indices = index.search(query_embedding, k)
    return [passages[i] for i in indices[0]]

# Choose a model
gen_model = genai.GenerativeModel('gemini-1.5-flash')

# Define the chatbot function with RAG
def chatbot(prompt, state, pdf_file):
    handle_pdf(pdf_file)  # This function does not modify state; it prepares data for retrieval
    retrieved_passages = retrieve_passages(prompt)
    augmented_prompt = prompt + "\n\n" + "\n\n".join(retrieved_passages)
    response = gen_model.generate_content(augmented_prompt)
    
    return response.text, state
```

```python
# Create Gradio interface
iface = gr.Interface(
    fn=chatbot,
    inputs=[gr.Textbox(label="Your Question"), gr.State(), gr.File(label="Upload PDF")],  # Updated input types
    outputs=[gr.Textbox(label="Response"), gr.State()],  # Updated output types
    title="Generative AI ChatBot",
    description="Hey! I'm your personal Generative AI ChatBot. Ask me anything based on the PDF you upload!",
)

# Launch the interface
iface.launch()
```

## Updated Task

```python
# Extract text from PDF but now with page numbers
def extract_text_from_pdf(pdf_path):
    document = fitz.open(pdf_path)
    passages = []
    for page_num in range(document.page_count):
        page = document.load_page(page_num)
        text = page.get_text()
        for passage in text.split("\n\n"):
            if passage.strip():
                passages.append((passage, page_num + 1))
    return passages

def handle_pdf(pdf_path):
    global passages, index, sentence_model
    passages = extract_text_from_pdf(pdf_path.name)
    texts = [p[0] for p in passages]
    sentence_model = SentenceTransformer('all-MiniLM-L6-v2')
    embeddings = sentence_model.encode(texts, convert_to_tensor=True)
    
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(embeddings.cpu().numpy())

def retrieve_passages(query, k=5, threshold=1.0):
    query_embedding = sentence_model.encode(query, convert_to_tensor=True)
    query_embedding = np.expand_dims(query_embedding, axis=0)
    distances, indices = index.search(query_embedding, k)
    
    relevant_passages = []
    for dist, idx in zip(distances[0], indices[0]):
        if dist < threshold:
            relevant_passages.append(passages[idx])
    
    return relevant_passages
```

```python

# Choose a model
gen_model = genai.GenerativeModel('gemini-1.5-flash')

# Define the chatbot function with RAG
def chatbot(prompt, state, pdf_file):
    handle_pdf(pdf_file)
    retrieved_passages = retrieve_passages(prompt)
    
    if not retrieved_passages:
        response_text = "I don't have this information. For more information, contact +123456789."
    else:
        context_text = ""
        page_numbers = set()
        for passage, page_num in retrieved_passages:
            context_text += f"{passage}\n\n"
            page_numbers.add(page_num)
        
        augmented_prompt = prompt + "\n\n" + context_text
        response = gen_model.generate_content(augmented_prompt)
        response_text = response.text

        # Format page numbers
        page_numbers_list = sorted(page_numbers)
        page_info = "The information is detailed on page(s) " + ", ".join(map(str, page_numbers_list)) + " of the PDF."
        response_text += "\n\n" + page_info
    
    return response_text, state

# Create Gradio interface
iface = gr.Interface(
    fn=chatbot,
    inputs=[gr.Textbox(label="Your Question"), gr.State(), gr.File(label="Upload PDF")],
    outputs=[gr.Textbox(label="Response"), gr.State()],
    title="Generative AI ChatBot",
    description="Hey! I'm your personal Generative AI ChatBot. Ask me anything based on the PDF you upload!",
)

# Launch the interface
iface.launch()
```
